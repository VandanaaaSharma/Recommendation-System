```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('/content/tmdb_5000_movies.csv')
credits = pd.read_csv('/content/tmdb_5000_credits.csv')
```

```python
movies.head(1)
```

```python
credits.head(1)
```

```python
# here wee merge the data of two datasets on the basis of title
movies = movies.merge(credits, on='title')
```

```python
movies.head(1)
```

```python
movies=movies[['movie_id','title','genres','overview','keywords','cast','crew']]
```

```python
movies.head()
```

```python
# // Missing data
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
# to check duplicate data is present or not
movies.duplicated().sum()
```

```python
# to make list in a proper way
import ast
def convert(obj):
  L=[]
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
# This is for cast we can no take the name of full cast instead of taking full we are going to take top 3 members name
import ast
def convert3(obj):
  L=[]
  counter = 0
  for i in ast.literal_eval(obj):
    if counter != 3:
      L.append(i['name'])
      counter +=1
    else:
      break
  return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
# Now we are going to preprocess the data according to crew there are many name from the crew and with their job title and name we need to select only director according to director we are going to recommend the movie
import ast
def fetch_director(obj):
  L=[]
  for i in ast.literal_eval(obj):
    if i['job']== 'Director':
      L.append(i['name'])
      break
  return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
#  Now we are gooing to work on overview column and convert it into list
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
# Make a column tags in which all the things combine given to us and when we get the tags column we remove all other column
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
new_df=movies[['movie_id','title','tags']]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
      y.append(ps.stem(i))
      return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
def recommend(movie):
  movie_index = new_df[new_df['title']==movie].index[0]
  distances = similarity[movie_index]
  movie_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
# print movies name
  print("Recommended movies for", movie, ":")  # Added a descriptive print
  for i in movie_list:
    recommended_movie_index = i[0]
    recommended_movie_title = new_df.iloc[recommended_movie_index]['title']  # Get title from DataFrame
    print(recommended_movie_title)
```

```python
recommend('Avatar')
```
